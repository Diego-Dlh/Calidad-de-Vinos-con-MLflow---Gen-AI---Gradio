```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../data/winequality-white.csv")
```

```python
print("Primeras filas del dataset:")
display(data.head())
```

```python
print("Descripción estadística:")
display(data.describe())
```

```python
print("Información del dataset:")
data.info()
```

```python
print("Valores nulos por columna:")
print(data.isnull().sum())
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="quality", data=data)
plt.title("Distribución Calidad de Vino")
plt.show()
```

```python
plt.figure(figsize=(12,8))
corr = data.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlación entre variables")
plt.show()
```

```python
corr_quality = corr["quality"].sort_values(ascending=False)
print("Variables más correlacionadas con calidad:")
print(corr_quality)
```

```python
variables_relevantes = corr_quality.index[1:6]  # top 5 ligadas a quality

for var in variables_relevantes:
    plt.figure(figsize=(6,4))
    sns.scatterplot(x=var, y="quality", data=data)
    plt.title(f"Relación {var} - calidad")
    plt.show()
```
